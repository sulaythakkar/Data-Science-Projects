```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv",encoding='ISO-8859-1')
star_wars.head(10)
```

```python
star_wars.columns
```

```python
yes_no = {
    "Yes": True,
    "No": False
}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
star_wars.head()
```

```python
cols_convert = star_wars.columns[3:9]
cols_map = {}
for col in cols_convert:
    if col == "Which of the following Star Wars films have you seen? Please select all that apply.":
        cols_map["Star Wars: Episode I  The Phantom Menace"] = True
    elif col == "Unnamed: 4":
        cols_map["Star Wars: Episode II  Attack of the Clones"] = True
    elif col == "Unnamed: 5":
        cols_map["Star Wars: Episode III  Revenge of the Sith"] = True
    elif col == "Unnamed: 6":
        cols_map["Star Wars: Episode IV  A New Hope"] = True
    elif col == "Unnamed: 7":
        cols_map["Star Wars: Episode V The Empire Strikes Back"] = True
    elif col == "Unnamed: 8":
        cols_map["Star Wars: Episode VI Return of the Jedi"] = True
    
    cols_map[np.NaN] = False
    star_wars[col] = star_wars[col].map(cols_map)    
```

```python
for col,i in zip(cols_convert, range(1,7)):
    star_wars.rename(columns={col: "seen_" + str(i)}, inplace=True)
print(star_wars.columns)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
for col,i in zip(star_wars.columns[9:15], range(1,7)):
    star_wars.rename(columns={col: "ranking_" + str(i)},inplace=True)
    
print(star_wars.columns[9:15])
```

```python
star_wars[star_wars.columns[9:15]].mean().sort_index()
```
